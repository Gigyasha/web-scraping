```python
from selenium import webdriver  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.chrome.options import Options 
import time
from bs4 import BeautifulSoup
import pandas as pd
import csv
from random import randint
from selenium.webdriver.support.wait import WebDriverWait
 
url="https://www.dawabazar.in/dealer/jaipur/385"
link=url
chrome_options = Options()  
chrome_options.add_argument("--headless")  
driver =webdriver.Chrome(executable_path=r'C:\Users\DELL\.wdm\drivers\chromedriver\80.0.3987.106\win32\chromedriver.exe')
driver.get(url)
time.sleep(5)

while True:
    try:
        morebox=driver.find_element_by_class_name("morebox")
        loadMoreButton=morebox.find_element_by_tag_name("a")
        time.sleep(4)
        loadMoreButton.click()
        time.sleep(4)
    except Exception as e:
        print(e)
        break
print("complete")
time.sleep(2)

data_link=[]


bl=driver.find_elements_by_class_name("cont_bg")
for b in bl:
    try:
        h1_tag=b.find_element_by_tag_name("h1")
        a_tag=h1_tag.find_element_by_tag_name("a").get_attribute("href")
        data_link.append(a_tag)
        print(a_tag)
    except Exception as e:
        print("error",e)
        continue


city="jaipur"
print(len(data_link))
head=["name","phone","mobile_no","address","city"]
with open('dawa_bazaar_jaipur.csv', mode='w+',encoding="UTF-8",newline="") as file:
    writer1 = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer1.writerow(head)
    for li in data_link:
        try:
            driver.get(li)
            name=None
            address=None
            phone_no=None
            mobile_no=None
            #cat=None
            time.sleep(3)
            block=driver.find_element_by_class_name("cont_bg")
            details_block=block.find_element_by_class_name("rerit")
            name=(details_block.find_element_by_tag_name("h1").text).strip()
            try:
                mobile_no=(details_block.find_element_by_class_name("pd_mobile").text).strip()
            except Exception as e:
                print("no mobile",e)
            try:
                phone_no=(details_block.find_element_by_class_name("pd_phone").text).strip()
            except Exception as e:
                print("no phone",e)
            address=(details_block.find_element_by_class_name("pd_address").text).strip()
            writer1.writerow([name,phone_no,mobile_no,address,city])
            print("NAME",name)
            print("Mobile",mobile_no)
            print("PHONE",phone_no)
            print("ADDRESS",address)
            print("City",city)

        except Exception as e:
            print("error",e)
            continue
```
