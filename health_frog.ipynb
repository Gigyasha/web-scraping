```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:


from selenium import webdriver  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.chrome.options import Options 
import time
from bs4 import BeautifulSoup
import pandas as pd
import csv
from random import randint
from selenium.webdriver.support.wait import WebDriverWait
 
url="https://www.healthfrog.in/chemists/list/rajasthan/jodhpur"
link=url
chrome_options = Options()  
chrome_options.add_argument("--headless")  
driver =webdriver.Chrome(executable_path=r'C:\Users\DELL\.wdm\drivers\chromedriver\80.0.3987.106\win32\chromedriver.exe')
driver.get(url)
time.sleep(5)
data_link=[]
while True:
    try:
        morebox=driver.find_element_by_class_name("pagination")
        loadMoreButton=morebox.find_element_by_tag_name("a")
        time.sleep(4)
        loadMoreButton.click()
        time.sleep(4)
    except Exception as e:
        print(e)
        break
print("complete")
time.sleep(2)
city="bikaner"
bl=driver.find_elements_by_class_name("listing ")
for b in bl:
    try:
        h3_tag=b.find_element_by_tag_name("h3")
        a_tag=h3_tag.find_element_by_tag_name("a").get_attribute("href")
        data_link.append(a_tag)
        print(a_tag)
    except Exception as e:
        print("error",e)
        continue
        
print(len(data_link))
bl=0
head=["name","phone","mobile_no","address","city"]
with open('dawa_bazaar_jodhpur.csv', mode='a+',encoding="UTF-8",newline="") as file:
    writer1 = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer1.writerow(head)
    for li in data_link:
            try:
                driver.get(li)
                name=None
                address=None
                phone=None
                mobile=None
                #cat=None
                time.sleep(5)
                title_tag=driver.find_element_by_class_name("title")
                name=(title_tag.find_element_by_tag_name("h1").text).strip()
                details_block=driver.find_element_by_class_name("listing")
                p_tag=details_block.find_elements_by_tag_name("p")
                for p in p_tag:
                    if "ndia" in p.text:
                        address=(p.text).strip()
                    if (p.text).isalnum() or "(" in (p.text) or "+" in (p.text) :
                        phone=(p.text).strip()
                        if "dit" in phone:
                            phone=phone.replace("Edit"," ",1)
                bl=bl+1
                print("block--",bl)
                writer1.writerow([name,phone,mobile,address,city])
                print("NAME",name)
                print("Mobile",mobile)
                print("PHONE",phone)
                print("ADDRESS",address)
                print("City",city)
            except Exception as e:
                print("skipp",e)
                continue

```
