#### Extracting companies data using selenium webdriver

```python
from selenium import webdriver  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.chrome.options import Options 
import time
from bs4 import BeautifulSoup
head=["Company_name","Email","Website","Address","Director1_name","Director1_designation","Director2_name","Director2_designation","Director3_name","Director3_designation","Director4_name","Director4_designation","Director5_name","Director5_designation"]   

id2=0
chrome_options = Options()  
chrome_options.add_argument("--headless")  
driver =webdriver.Chrome(executable_path=r'C:\Users\DELL\.wdm\drivers\chromedriver\80.0.3987.106\win32\chromedriver.exe')
bl=0
with open('dipp_comp_3k_4k.csv', mode='a+',encoding="UTF-8",newline="") as file:
    writer1 = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer1.writerow(head)

    for comp_name in full_comp_list[3505:4000]: 
        try:
            url="https://www.zaubacorp.com/companysearchresults/"+comp_name
            driver.get(url)
            time.sleep(4)
            driver.find_element_by_xpath('//*[@id="results"]/tbody/tr[1]/td[2]').click()
            time.sleep(4)
            try:
                driver.find_element_by_xpath('//*[@id="cbb"]/svg').click()
                time.sleep(2)
                try:
                    driver.find_element_by_xpath('//*[@id="report_text"]').click()
                except Exception as e:
                    print("no stop google add",e)
            except Exception as e:
                print("NO add",e)
                
            details=driver.find_element_by_xpath('//*[@id="block-system-main"]/div[2]/div[1]/div[5]/div/div[1]')
            #//*[@id="block-system-main"]/div[2]/div[1]/div[5]/div/div[1]
            #//*[@id="block-system-main"]/div[2]/div[1]/div[5]/div/div[1]
            complete_det=details.text
            val=complete_det.splitlines( )
            #print(val)
            email=None
            address=None
            website=None
            ind_em=val[0].find(":")
            ind_web=val[1].find(":")
            ind_add=val.index("Address:")
            email=(val[0])[ind_em+1:].strip()
            website=val[1].strip()
            address=val[ind_add+1]
            l_contact=[]
            if "Click here" in website:
                website=None
            else:
                website=None
            l_contact.append(email) 
            l_contact.append(website)
            l_contact.append(address)
            #print(l_contact)
            time.sleep(4)
            tab_head=driver.find_elements_by_xpath('//*[@id="block-system-main"]/div[2]/div[1]/div[7]/table/thead/tr/td')
            #tab_head=tab_head.splitlines( )
            l=[]
            for tab in tab_head:
                l.append(tab.text)
            ind_name=l.index("Director Name")
            ind_deg=l.index("Designation")
            time.sleep(4)
            dir_data_list=driver.find_elements_by_class_name('accordion-toggle')   #row wise data from tble
            #print(dir_data_list)
            j=1
            dir_dict={"dir1":[None,None],"dir2":[None,None],"dir3":[None,None],"dir4":[None,None],"dir5":[None,None]}
            for i in dir_data_list:
                t=[]
                k=(i.text).splitlines( )
                #t.append(j)
                t.append(k[ind_name])
                t.append(k[ind_deg])
                dir_dict["dir"+str(j)]=t
                j=j+1
            #print(dir_dict)
            dir_data=[]
            for s in dir_dict.values():
                dir_data.extend(s)
            #print(dir_data)
            #l_contact.extend(dir_data)
            bl=bl+1
            print("BLOCK :",bl,"COMPLETED")
            final_list=[]
            final_list.append(comp_name)
            final_list.extend(l_contact)
            final_list.extend(dir_data)
            print(final_list)
            writer1.writerow(final_list)
        except Exception as e:
            print("OOPS THIS DOES NOT EXIST",e)
            writer1.writerow([comp_name,None,None,None,None,None,None,None,None,None,None,None,None,None])
            continue
```
